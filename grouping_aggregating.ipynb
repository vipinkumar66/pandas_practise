```python
df['Gender'].count()
```

```python
# This tell us that majority of the survey was answered by the Man
df['Gender'].value_counts()
```

If you see the above two cell you will understand that the df.value_counts() count how much time each individual has occured where as the df.count() counts the total non none values.

```python
# And now if you want to get the above data in the percentage format than we can do it like this
df['Gender'].value_counts(normalize=True)
# So around 91% are Men
```

Now the above gives the Gender with respect to the whole dataframe but what if we want it country wise. So here comes the use of the Grouping

```python
df['Country'].value_counts()
```

```python
country_grp = df.groupby("Country")
```

```python
country_grp.get_group("India")
```

If you see the above data it is from the India, so the group by function here acted as a filter.
filter = df['Country] == 'India'
df.loc[filt] => this thing is going to give the same result

```python
#  Now there is no use to run filter each time for different countries, just change the name and get the output
country_grp["Gender"].value_counts(normalize=True).loc['India']
```
