```python
import pathlib
import pandas as pd
import numpy as np

BASE_DIR = pathlib.Path().resolve().parent
pandas = BASE_DIR / "pandas_deep_dive"
csv1 = pandas / "survey_results_schema.csv"
csv2 = pandas / "survey_results_public.csv"

print(BASE_DIR, csv1.exists(), csv2.exists())
```

```python
people_data = {
    'first':['vipin', 'varun', 'asrh','mahesh', np.nan, None, 'missing'],
    'last':['kumar', np.nan, 'dhiman', 'singh', 'missing', None, None],
    'age':[24, 24, 'Missing', 30, np.nan, None, None],
    'email':['vipin@gmail.com', 'varun@gmail.com', 'arsh@gmail.com', None, 'missing', 'missing','missing']
}
# since here we are creating our dataframe with the dictionary so we can replace it while loading only
df = pd.DataFrame(people_data)
df.replace("missing", np.nan, inplace=True)
df.replace("Missing", np.nan, inplace=True)
```

```python
df
```

```python
# so first lets drop the missing values
df.dropna()
```

```python
# The above one gives only 2 values and rest all are removed if we see the arguments paased to the dropna are:
df.dropna(axis='index', how='any')
```

As you can see the above result match the previous output. Here any means that if there is any empty value in the row (as we have set the axis to index to set it to column we have to pass column to index) it will delete that entire row.
If we make it to all than it will check if in a row all the values are missing or not. If all the values are missing than only it will delete that row otherwise not

```python
df.dropna(axis='index', how='all')
```

```python
# here i am saying that if in email column value is missing just remove that row
df.dropna(axis='index', how='any',subset=['email'])
# so here it removed the mahesh row
```

```python
df.dropna(how='any',subset=['first','email'])
```

```python
# This will tell which all values are empty
df.isna()
```

```python
# Suppose you want to fill all the missing values this is mostly good for the numerical data
df.fillna("MISSING")
```

```python
na_vals = ['NA', 'Na', 'Missing']
csv_df = pd.read_csv(csv2, na_values=na_vals) #this is how we deal with the missing values
csv_df
```

```python
csv_df.columns
```

```python
csv_df['WorkExp'].head(20)
# since most of the values which are seeing are empty so it is going to have a negative affect on our data
```

```python
csv_df['WorkExp'].mean()
```

```python
csv_df['new_exp'] = csv_df['WorkExp'].fillna(csv_df['WorkExp'].mean())
csv_df['new_exp'].mean()
# As you can see there was not much difference, so this is sometime considered a good approach
```
