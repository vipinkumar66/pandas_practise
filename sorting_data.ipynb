```python
import pathlib
import pandas as pd

BASE_DIR = pathlib.Path().resolve().parent
pandas = BASE_DIR / "pandas_deep_dive"
csv1 = pandas / "survey_results_schema.csv"
csv2 = pandas / "survey_results_public.csv"

print(BASE_DIR, csv1.exists(), csv2.exists())
```

```python
people = {
    "name":["vipin", "varun", "mashesh"],
    "age": [24, 24, 27],
    "email":["vipin@gmail.com", "varun@gmail.com", "mahesh@yahoo.com"]
}
eg_df = pd.DataFrame(people)
eg_df
```

#Sorting on a single column
The sorting method is always in the ascending order and if you want to do it in descending than pass ascending=False

```python
eg_df.sort_values(by="name")
```

```python
eg_df.sort_values(by="name", ascending=False)
```

#Multiple Columns

Why we go for the multiple columns?
Suppose we have some columns which has same values so sorting them on one column may not give the proper result so that time we can take multiple columns.
First it will sort on the starting index of the columns list and than the next one.

```python
eg_df = eg_df.append({"name":"Arshita", "age":21, "email":"arshita@gmail.com"}, ignore_index=True)
```

```python
eg_df.sort_values(by=["email","name"], ascending=False)
```

```python
# In the above cell we are taking the descending for the both but what if i want to take ascending for the first and the descending for the second
eg_df.loc[1, ["age"]] = 25
eg_df.sort_values(by=["name", "age"], ascending=[True, False], inplace=True)

eg_df
```

As you can see in the above output it arranged all values in the ascending order with respect to name and than the varun and vipn they both were starting with the same word so here it goes with respect to age and we have give descending order for age so it took the varun first as his age was 25

```python
# So now if we want the order back to the previous version than we can sort it on index
eg_df.sort_index()
```

```python
#  WE can sort individual columns also
eg_df["name"].sort_values()
```
